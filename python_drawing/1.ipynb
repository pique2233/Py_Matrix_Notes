```python
import numpy as np
import matplotlib.pyplot as plt

# 
labels = ['Motion Binding', 'Action Binding', 'Object Interactions', 'Spatial Relationships', 
          'Dynamic Attr', 'Consis Attr', 'Generative Numeracy']

# 
LVD = [0.75, 0.70, 0.80, 0.85, 0.90, 0.80, 0.65]
T2V_Turbo_V2 = [0.85, 0.80, 0.85, 0.75, 0.85, 0.90, 0.70]
ModelScope = [0.80, 0.75, 0.75, 0.80, 0.80, 0.85, 0.80]
VideoCrafter2 = [0.70, 0.85, 0.80, 0.70, 0.75, 0.75, 0.90]
Gen_2 = [0.60, 0.65, 0.75, 0.70, 0.80, 0.85, 0.85]
Gen_3 = [0.90, 0.88, 0.90, 0.92, 0.95, 0.93, 0.80]
CogVideoX_5B = [0.85, 0.80, 0.78, 0.86, 0.92, 0.90, 0.87]

# 
def normalize(data):
    return [(x - min(data)) / (max(data) - min(data)) for x in data]

# 
LVD = normalize(LVD)
T2V_Turbo_V2 = normalize(T2V_Turbo_V2)
ModelScope = normalize(ModelScope)
VideoCrafter2 = normalize(VideoCrafter2)
Gen_2 = normalize(Gen_2)
Gen_3 = normalize(Gen_3)
CogVideoX_5B = normalize(CogVideoX_5B)

# 
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

# 
LVD += LVD[:1]
T2V_Turbo_V2 += T2V_Turbo_V2[:1]
ModelScope += ModelScope[:1]
VideoCrafter2 += VideoCrafter2[:1]
Gen_2 += Gen_2[:1]
Gen_3 += Gen_3[:1]
CogVideoX_5B += CogVideoX_5B[:1]
angles += angles[:1]

# 
colors = {
    'LVD': '#CB83CC',
    'T2V-Turbo-V2': '#BDAEE8',
    'ModelScope': '#AFD1FA',
    'VideoCrafter2': '#D1E3D8',
    'Gen-2': '#F6E9CB',
    'Gen-3': '#8A7A92',  # 
    'CogVideoX-5B': '#6C5A7B'  # 
}

#  dpi
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True), dpi=300)

# 
ax.plot(angles, LVD, color=colors['LVD'], linewidth=2, linestyle='-', label='LVD')
ax.fill(angles, LVD, color=colors['LVD'], alpha=0.25)

ax.plot(angles, T2V_Turbo_V2, color=colors['T2V-Turbo-V2'], linewidth=2, linestyle='-', label='T2V-Turbo-V2')
ax.fill(angles, T2V_Turbo_V2, color=colors['T2V-Turbo-V2'], alpha=0.25)

ax.plot(angles, ModelScope, color=colors['ModelScope'], linewidth=2, linestyle='-', label='ModelScope')
ax.fill(angles, ModelScope, color=colors['ModelScope'], alpha=0.25)

ax.plot(angles, VideoCrafter2, color=colors['VideoCrafter2'], linewidth=2, linestyle='-', label='VideoCrafter2')
ax.fill(angles, VideoCrafter2, color=colors['VideoCrafter2'], alpha=0.25)

ax.plot(angles, Gen_2, color=colors['Gen-2'], linewidth=2, linestyle='--', label='Gen-2')
ax.fill(angles, Gen_2, color=colors['Gen-2'], alpha=0.25)

ax.plot(angles, Gen_3, color=colors['Gen-3'], linewidth=2, linestyle='--', label='Gen-3')
ax.fill(angles, Gen_3, color=colors['Gen-3'], alpha=0.25)

ax.plot(angles, CogVideoX_5B, color=colors['CogVideoX-5B'], linewidth=2, linestyle='-', label='CogVideoX-5B')
ax.fill(angles, CogVideoX_5B, color=colors['CogVideoX-5B'], alpha=0.25)

# 
ax.set_yticklabels([])

# 
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=8, fontweight='bold', color='black', ha='center')

# 
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# 
plt.title('(c) Benchmarking T2V Models', size=16)

# 
plt.show()

```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# 
text = """
Python is a powerful programming language. It is widely used in web development, data analysis, artificial intelligence, and more.
Data science is a field that uses scientific methods, processes, algorithms and systems to extract knowledge and insights from structured and unstructured data.
Machine learning is a subset of artificial intelligence that provides systems the ability to learn and improve from experience without being explicitly programmed.
"""

# 
mask = np.array(Image.open("mask_image.png"))  # 

# 
wordcloud = WordCloud(
    width=1200,
    height=800,
    background_color='white',
    mask=mask,  # 
    contour_width=2,  # 
    contour_color='steelblue',  # 
    colormap='plasma',  # 
    max_words=100,  # 
    max_font_size=200,  # 
    min_font_size=10,  # 
    scale=3,  # 
    stopwords=STOPWORDS.union({"used", "field", "systems", "ability"}),  # 
    font_path=None,  # 
    prefer_horizontal=0.9  # 
).generate(text)

# 
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 
plt.title('Python, Data Science, and Machine Learning Word Cloud', fontsize=24)

# 
plt.savefig('wordcloud_python_data_science_machine_learning_custom.png')

# 
plt.show()
```
